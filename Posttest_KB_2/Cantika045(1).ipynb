```python
import pandas as pd
from prettytable import PrettyTable
from tabulate import tabulate
df = pd.read_csv("test.csv")
df.info()
```

Informasi yang didapatkan adalah jumlah record, nama attribute, jumlah attribute dengan tipe objek, jumlah attribute dengan tipe angka, jumlah attribute yang memiliki nilai null.

```python
#LihatJumlahRecord
Jumlah = len(df)
print(Jumlah)

#LihatNamaColumns
nama = list(df.columns)
print(nama)

#LihatJumlahAtrTipeObjek
jumlahObjek = len(df.dtypes[df.dtypes == 'object'])
print(jumlahObjek)

#LihatJumlahAtrTipeAngka
jumlahAngka =len(df.dtypes[df.dtypes == 'int64'])
print(jumlahAngka)

#LihatJumlahAtrNilaiNull
df.isna().any().sum()
```

Masukkan informasi di atas dalam dictionary dengan key sesuai nama informasinya. Tampilkan isi dari dictionary tersebut sebagus mungkin.

```python
Jumlah = len(df)

nama = list(df.columns)
nama = ",\n".join(nama)

jumlahObjek = len(df.dtypes[df.dtypes == 'object'])


jumlahAngka = len(df.dtypes[df.dtypes == 'int64'])

jumlahNull = df.isna().any().sum()


kamus = {
    "Jumlah Record":[Jumlah],
    "Nama Atribut" : [nama],
    "Jumlah atribut tipe objek" : [jumlahObjek],
    "Jumlah atribut tipe angka" : [jumlahAngka],
    "Jumlah atribut nilai null"  : [jumlahNull],
    
}

tb = pd.DataFrame(kamus)
print(tabulate(tb, headers='keys', tablefmt='psql',showindex=False))
```
